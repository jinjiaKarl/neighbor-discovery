```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

df_ed25519_blake2b_receiver = pd.read_csv('./data/ECC/ed25519/receiver_improved_protocol_ed25519_BLAKE2b.csv')
df_ed25519_md5_receiver = pd.read_csv('./data/ECC/ed25519/receiver_improved_protocol_ed25519_MD5.csv')
df_ed25519_sha256_receiver = pd.read_csv('./data/ECC/ed25519/receiver_improved_protocol_ed25519_SHA256.csv')
df_ed25519_sha3_256_receiver = pd.read_csv('./data/ECC/ed25519/receiver_improved_protocol_ed25519_SHA3_256.csv')
df_ed25519_blake2b_transmitter = pd.read_csv('./data/ECC/ed25519/transmitter_improved_protocol_ed25519_BLAKE2b.csv')
df_ed25519_md5_transmitter = pd.read_csv('./data/ECC/ed25519/transmitter_improved_protocol_ed25519_MD5.csv')
df_ed25519_sha256_transmitter = pd.read_csv('./data/ECC/ed25519/transmitter_improved_protocol_ed25519_SHA256.csv')
df_ed25519_sha3_256_transmitter = pd.read_csv('./data/ECC/ed25519/transmitter_improved_protocol_ed25519_SHA3_256.csv')

df_ed448_blake2b_receiver = pd.read_csv('./data/ECC/ed448/receiver_improved_protocol_ed448_BLAKE2b.csv')
df_ed448_md5_receiver = pd.read_csv('./data/ECC/ed448/receiver_improved_protocol_ed448_MD5.csv')
df_ed448_sha256_receiver = pd.read_csv('./data/ECC/ed448/receiver_improved_protocol_ed448_SHA256.csv')
df_ed448_sha3_256_receiver = pd.read_csv('./data/ECC/ed448/receiver_improved_protocol_ed448_SHA3_256.csv')
df_ed448_blake2b_transmitter = pd.read_csv('./data/ECC/ed448/transmitter_improved_protocol_ed448_BLAKE2b.csv')
df_ed448_md5_transmitter = pd.read_csv('./data/ECC/ed448/transmitter_improved_protocol_ed448_MD5.csv')
df_ed448_sha256_transmitter = pd.read_csv('./data/ECC/ed448/transmitter_improved_protocol_ed448_SHA256.csv')
df_ed448_sha3_256_transmitter = pd.read_csv('./data/ECC/ed448/transmitter_improved_protocol_ed448_SHA3_256.csv')

# Concatenate receiver and transmitter DataFrames for each key
def concatenate_receiver_transmitter(receiver_df, transmitter_df):
    receiver_df['Source'] = 'Receiver'
    transmitter_df['Source'] = 'Receiver'
    return pd.concat([receiver_df, transmitter_df], ignore_index=True)

# Concatenate for each key
df_concatenated_ed25519_blake2b = concatenate_receiver_transmitter(df_ed25519_blake2b_receiver, df_ed25519_blake2b_transmitter)
df_concatenated_ed25519_md5 = concatenate_receiver_transmitter(df_ed25519_md5_receiver, df_ed25519_md5_transmitter)
df_concatenated_ed25519_sha256 = concatenate_receiver_transmitter(df_ed25519_sha256_receiver, df_ed25519_sha256_transmitter)
df_concatenated_ed25519_sha3_256 = concatenate_receiver_transmitter(df_ed25519_sha3_256_receiver, df_ed25519_sha3_256_transmitter)
df_concatenated_ed448_blake2b = concatenate_receiver_transmitter(df_ed448_blake2b_receiver, df_ed448_blake2b_transmitter)
df_concatenated_ed448_md5 = concatenate_receiver_transmitter(df_ed448_md5_receiver, df_ed448_md5_transmitter)
df_concatenated_ed448_sha256 = concatenate_receiver_transmitter(df_ed448_sha256_receiver, df_ed448_sha256_transmitter)
df_concatenated_ed448_sha3_256 = concatenate_receiver_transmitter(df_ed448_sha3_256_receiver, df_ed448_sha3_256_transmitter)

# Choose a column to compare
column_to_compare = 'Total Time'
scientific_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22']
# Extract data for the chosen column from the concatenated DataFrames
data_columns_concatenated = {
    'ed25519_blake2b': df_concatenated_ed25519_blake2b[column_to_compare],
    'ed25519_md5': df_concatenated_ed25519_md5[column_to_compare],
    'ed25519_sha256': df_concatenated_ed25519_sha256[column_to_compare],
    'ed25519_sha3_256': df_concatenated_ed25519_sha3_256[column_to_compare],
    'ed448_blake2b': df_concatenated_ed448_blake2b[column_to_compare],
    'ed448_md5': df_concatenated_ed448_md5[column_to_compare],
    'ed448_sha256': df_concatenated_ed448_sha256[column_to_compare],
    'ed448_sha3_256': df_concatenated_ed448_sha3_256[column_to_compare],
}

# Calculate mean and standard deviation
means = [data_column.mean() for data_column in data_columns_concatenated.values()]
std_devs = [data_column.std() for data_column in data_columns_concatenated.values()]
n_values = [len(data_column) for data_column in data_columns_concatenated.values()]

# Calculate 95% confidence interval
ci = [1.96 * std_dev / np.sqrt(n) for std_dev, n in zip(std_devs, n_values)]

# Prepare data for plotting
keys = list(data_columns_concatenated.keys())
x_pos = np.arange(len(keys))

# Plot bar graph with mean and standard deviation
plt.figure(figsize=(10, 6))
bars = plt.bar(x_pos, means, yerr=ci, align='center', alpha=0.7, ecolor='black', capsize=10, width=0.5, color=scientific_colors, edgecolor='black')

# Customize legend
legend_elements = [Patch(facecolor=color, label=key) for key, color in zip(keys, scientific_colors)]
plt.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(1, 1), title='Keys', frameon=False)

# Remove x-axis labels
plt.xticks([])

# Add data labels on top of the bars
for i, mean in enumerate(means):
    plt.text(x_pos[i], mean + 0.02 * max(means), f'{mean:.3f}', ha='center', va='bottom', fontsize=9)

plt.xlabel('Keys', fontsize=12)
plt.ylabel('Time (s)', fontsize=12)
plt.title(f'Total time in Improved Protocol at 95% CI', fontsize=14)

# Remove the top and right spines for aesthetics
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```
