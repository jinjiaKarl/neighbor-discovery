```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

df_AES_16_1024_receiver = pd.read_csv('./data/AES_16/1024/receiver_basic_protocol_updated.csv')
df_AES_16_2048_receiver = pd.read_csv('./data/AES_16/2048/receiver_basic_protocol_updated.csv')
df_AES_16_3072_receiver = pd.read_csv('./data/AES_16/3072/receiver_basic_protocol_updated.csv')
df_AES_16_1024_transmitter = pd.read_csv('./data/AES_16/1024/transmitter_basic_protocol_updated.csv')
df_AES_16_2048_transmitter = pd.read_csv('./data/AES_16/2048/transmitter_basic_protocol_updated.csv')
df_AES_16_3072_transmitter = pd.read_csv('./data/AES_16/3072/transmitter_basic_protocol_updated.csv')
df_AES_24_1024_receiver = pd.read_csv('./data/AES_24/1024/receiver_basic_protocol_updated.csv')
df_AES_24_2048_receiver = pd.read_csv('./data/AES_24/2048/receiver_basic_protocol_updated.csv')
df_AES_24_3072_receiver = pd.read_csv('./data/AES_24/3072/receiver_basic_protocol_updated.csv')
df_AES_24_1024_transmitter = pd.read_csv('./data/AES_24/1024/transmitter_basic_protocol_updated.csv')
df_AES_24_2048_transmitter = pd.read_csv('./data/AES_24/2048/transmitter_basic_protocol_updated.csv')
df_AES_24_3072_transmitter = pd.read_csv('./data/AES_24/3072/transmitter_basic_protocol_updated.csv')
df_AES_32_1024_receiver = pd.read_csv('./data/AES_32/1024/receiver_basic_protocol_updated.csv')
df_AES_32_2048_receiver = pd.read_csv('./data/AES_32/2048/receiver_basic_protocol_updated.csv')
df_AES_32_3072_receiver = pd.read_csv('./data/AES_32/3072/receiver_basic_protocol_updated.csv')
df_AES_32_1024_transmitter = pd.read_csv('./data/AES_32/1024/transmitter_basic_protocol_updated.csv')
df_AES_32_2048_transmitter = pd.read_csv('./data/AES_32/2048/transmitter_basic_protocol_updated.csv')
df_AES_32_3072_transmitter = pd.read_csv('./data/AES_32/3072/transmitter_basic_protocol_updated.csv')

# Concatenate receiver and transmitter DataFrames for each key
def concatenate_receiver_transmitter(receiver_df, transmitter_df):
    receiver_df['Source'] = 'Receiver'
    transmitter_df['Source'] = 'Transmitter'
    return pd.concat([receiver_df, transmitter_df], ignore_index=True)

# Concatenate for each key
df_concatenated_1024_AES_16 = concatenate_receiver_transmitter(df_AES_16_1024_receiver, df_AES_16_1024_transmitter)
df_concatenated_2048_AES_16 = concatenate_receiver_transmitter(df_AES_16_2048_receiver, df_AES_16_2048_transmitter)
df_concatenated_3072_AES_16 = concatenate_receiver_transmitter(df_AES_16_3072_receiver, df_AES_16_3072_transmitter)
df_concatenated_1024_AES_24 = concatenate_receiver_transmitter(df_AES_24_1024_receiver, df_AES_24_1024_transmitter)
df_concatenated_2048_AES_24 = concatenate_receiver_transmitter(df_AES_24_2048_receiver, df_AES_24_2048_transmitter)
df_concatenated_3072_AES_24 = concatenate_receiver_transmitter(df_AES_24_3072_receiver, df_AES_24_3072_transmitter)
df_concatenated_1024_AES_32 = concatenate_receiver_transmitter(df_AES_32_1024_receiver, df_AES_32_1024_transmitter)
df_concatenated_2048_AES_32 = concatenate_receiver_transmitter(df_AES_32_2048_receiver, df_AES_32_2048_transmitter)
df_concatenated_3072_AES_32 = concatenate_receiver_transmitter(df_AES_32_3072_receiver, df_AES_32_3072_transmitter)

# Choose a column to compare
column_to_compare = 'Total time'
scientific_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22']
# Extract data for the chosen column from the concatenated DataFrames
data_columns_concatenated = {
    'AES_16_1024': df_concatenated_1024_AES_16[column_to_compare],
    'AES_16_2048': df_concatenated_2048_AES_16[column_to_compare],
    'AES_16_3072': df_concatenated_3072_AES_16[column_to_compare],
    'AES_24_1024': df_concatenated_1024_AES_24[column_to_compare],
    'AES_24_2048': df_concatenated_2048_AES_24[column_to_compare],
    'AES_24_3072': df_concatenated_3072_AES_24[column_to_compare],
    'AES_32_1024': df_concatenated_1024_AES_32[column_to_compare],
    'AES_32_2048': df_concatenated_2048_AES_32[column_to_compare],
    'AES_32_3072': df_concatenated_3072_AES_32[column_to_compare],
}

# Calculate mean and standard deviation
means = [data_column.mean() for data_column in data_columns_concatenated.values()]
std_devs = [data_column.std() for data_column in data_columns_concatenated.values()]
n_values = [len(data_column) for data_column in data_columns_concatenated.values()]

# Calculate 95% confidence interval
ci = [1.96 * std_dev / np.sqrt(n) for std_dev, n in zip(std_devs, n_values)]

# Prepare data for plotting
keys = list(data_columns_concatenated.keys())
x_pos = np.arange(len(keys))

# Plot bar graph with mean and standard deviation
plt.figure(figsize=(10, 6))
bars = plt.bar(x_pos, means, yerr=ci, align='center', alpha=0.7, ecolor='black', capsize=10, width=0.5, color=scientific_colors, edgecolor='black')

# Customize legend
legend_elements = [Patch(facecolor=color, label=key) for key, color in zip(keys, scientific_colors)]
plt.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(1, 1), title='Keys', frameon=False)

# Remove x-axis labels
plt.xticks([])

# Add data labels on top of the bars
for i, mean in enumerate(means):
    plt.text(x_pos[i], mean + 0.02 * max(means), f'{mean:.3f}', ha='center', va='bottom', fontsize=9)

plt.xlabel('Keys', fontsize=12)
plt.ylabel('Time (s)', fontsize=12)
plt.title(f'Total time in Basic Protocol at 95% CI', fontsize=14)

# Remove the top and right spines for aesthetics
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```
